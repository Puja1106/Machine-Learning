```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix
```

```python
df=pd.read_csv('bank-full.csv',sep=';')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.shape
```

```python
df= pd.get_dummies(df, columns=['job','marital','education','contact','month','poutcome', 'default', 'housing', 'loan','y'], drop_first=True)
df
```

```python
df.info()
```

```python
x=df.iloc[:,0:43]
y= df['y_yes'] 
```

```python
x
```

```python
y
```

# Split the dat into training and testing dataset


```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.70,random_state=142)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

# Build and train the model 


```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
ytrain_predict=model.predict(x_train)
ytrain_predict
```

```python
model.predict_proba(x_train)
```

#### Calculate the metrics for TRAINING DATA

```python
confusion_matrix(y_train,ytrain_predict)
```

```python
print(classification_report(y_train,ytrain_predict))
```
